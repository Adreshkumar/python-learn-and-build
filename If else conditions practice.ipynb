```python
pin = int(input("Enter PIN: "))

if pin == 1234:
    option = int(input("Choose 1 for balance check, 2 to withdraw money: "))

    if option == 1:
        print("Your current balance is 1000")

    elif option == 2:
        amount = int(input("Enter amount you want to withdraw: "))

        if 500 <= amount <= 1000:
            print("Thanks for the transaction")
        elif amount < 500:
            print("Minimum withdrawal is 500")
        else:
            print("Insufficient balance")

    else:
        print("Invalid option")

else:
    print("Invalid PIN")
```
