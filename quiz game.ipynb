```python
def new_game():
    """Starts a new quiz game session."""
    guesses = []
    correct_guesses = 0
    question_num = 0

    for question_text, correct_answer in questions.items():
        print("-------------------------")
        print(question_text)
        # Display options for the current question
        for option in options[question_num]:
            print(option)

        guess = input("Enter (A, B, C, or D): ").upper()
        guesses.append(guess)

        if guess == correct_answer:
            print("CORRECT!")
            correct_guesses += 1
        else:
            print(f"WRONG! The correct answer was {correct_answer}")
        
        question_num += 1

    display_score(correct_guesses, guesses)

def display_score(correct_guesses, guesses):
    """Displays the final results of the game."""
    print("-------------------------")
    print(" RESULTS ")
    print("-------------------------")

    print("Answers: ", end="")
    for answer in questions.values():
        print(answer, end=" ")
    print()

    print("Guesses: ", end="")
    for guess in guesses:
        print(guess, end=" ")
    print()

    score = int((correct_guesses / len(questions)) * 100)
    print(f"Your score is: {score}%")

def play_again():
    """Asks the user if they want to play another game."""
    response = input("Do you want to play again? (yes or no): ").upper()
    if response == "YES":
        return True
    else:
        return False

# --- Question Data ---
questions = {
    "Who created Python?: ": "A",
    "What year was Python created?: ": "B",
    "Python is tributed to which comedy group?: ": "C",
    "Is the Earth round?: ": "A"
}
# Options list must align with the order of questions
options = [
    ["A. Guido van Rossum", "B. Elon Musk", "C. Bill Gates", "D. Mark Zuckerburg"],
    ["A. 1989", "B. 1991", "C. 2000", "D. 2016"],
    ["A. Lonely Island", "B. Smosh", "C. Monty Python", "D. SNL"],
    ["A. True", "B. False", "C. sometimes", "D. What's Earth?"]
]

# --- Main Game Loop ---
if __name__ == "__main__":
    new_game()
    while play_again():
        new_game()
    print("Byeeeeee!")
```
