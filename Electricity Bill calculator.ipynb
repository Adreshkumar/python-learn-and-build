```python
def Electricity_Bill():
    print("Enter Your Units")
    Units_num=int(input())
    Bill=0
    if Units_num<=100:
        Bill=Units_num*5
    elif Units_num<=300:
        Bill=(100*5)+((Units_num-100)*8)
    else:
        Bill=(100*5)+(200*8)+((Units_num-300)*12)
    print(f"Your Total Bill: ${Bill}")
    
```
